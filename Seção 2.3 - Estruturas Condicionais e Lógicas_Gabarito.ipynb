## **Definição: Estruturas Condicionais e Lógicas**

As declarações condicionais, frequentemente chamadas de declarações if-then, permitem que executemos um trecho de código com base em alguma condição (uma condição booleana). As declarações condicionais em Python incluem:



*   **Condicional *if***
*   **Condicional *if-else***
*   **Condicional *if* (nested)**
*   **Condicional *if-elif-else***



### Exemplo de Trabalho: Tanque de Armazenamento de GLP

## **2.3.1 Condicional *if***

<p align="justify">Se a expressão da instrução "if" for avaliada como verdadeira, o código indentado que segue a declaração é executado. Se a expressão for avaliada como falsa, o código indentado que segue a instrução "if" é ignorado, e o programa executa a próxima linha de código que está indentada no mesmo nível da instrução "if".



image.png

```python
# Exemplo de condicional if

# Tpressure = 10 # em bar
Tpressure = int(input("Qual o valor da pressão no tanque em bar? "))
Tpressure_max = 50 # em bar

if Tpressure >= 50:
  print("\n O valor da pressão do tanque está fora ou é igual a pressão máxima permitida!")

if Tpressure < 50:
  print("\n O valor da pressão do tanque está dentro da pressão máxima permitida!")
```

## **2.3.2 Condicional *if-else***

<p align="justify">Esta declaração nos permite adicionar uma segunda possibilidade ao script, ou seja, o que fazer quando a condição for falsa. Vamos visualizar esse caso.

image.png

<p align="justify">O código indentado da instrução "if" é executado se a expressão for avaliada como verdadeira. O código indentado imediatamente após o "else" é executado apenas se a expressão for avaliada como falsa. Para marcar o fim do bloco "else", o código deve ser desindentado para o mesmo nível que a linha de início da instrução "if".

```python
# Exemplo para a condicional if-else

# Tpressure = 10 # em bar
Tpressure = int(input("Qual o valor da pressão no tanque em bar? "))
Tpressure_max = 50 # em bar

if Tpressure >= 50:
  print("\n O valor da pressão do tanque está fora ou é igual a pressão máxima permitida!"
        "\n Acionar protocolo de evacuação da área")
else:
  print("\n O valor da pressão do tanque está dentro da pressão máxima permitida!")
```

## **2.3.4 Condicional *if-elif-else***

A palavra-chave "elif" pode ser pensada como "senão, se". A utilizamos quando desejamos uma divisão mais distintiva entre o "if" e o "else". A instrução "elif" em Python permite que verificações adicionais sejam realizadas após uma instrução "if" inicial. Uma instrução "elif" difere da instrução "else" porque outra expressão é fornecida para ser verificada, da mesma forma que na instrução "if" inicial.

image.png


Se a expressão for verdadeira, o código indentado que segue a instrução "elif" é executado. Se a expressão for avaliada como falsa, o código pode continuar com uma instrução "else" opcional. Múltiplas instruções "elif" podem ser usadas após um "if" inicial para realizar uma série de verificações. Uma vez que uma expressão "elif" seja avaliada como verdadeira, nenhuma outra instrução "elif" ou a instrução "else" será executada.

```python
# Exemplo para a condicional if-elif-else

# Tpressure = 51 #bar
Tpressure = int(input("Qual o valor da pressão no tanque em bar? "))
Tpressure_max = 50 # em bar

if Tpressure > 50:
  print("\n O valor da pressão do tanque está fora da pressão máxima permitida!"
        "\n Acionar protocolo de evacuação da área!")
elif 50 <= Tpressure > 45:
  print("\n O valor da pressão do tanque está dentro da pressão máxima permitida")
  print("\n Entretanto, a pressão está na faixa de pressão que requer uma atenção")
else:
  print("\n O valor da pressão do tanque está dentro da pressão máxima permitida")
  print("\n Mantenham o bom trabalho!")
```

## **Definição: Condicionais com operadores lógicos**

Os operadores lógicos são utilizados para combinar múltiplas condições:

\\

*   **or (ou)** $→$ O operador é usado para verificar se pelo menos uma das condições é satisfeita;
*   **and (e)** $→$ O operador é usado para verificar se todas as condições são satisfeitas.
*   **in (dentro)** $→$ O operador é usado para verificar se verificar se um valor está presente em uma sequência.

\\

Vamos analisar a sintaxe e a aplicação dos operadores lógicos com base no mesmo exemplo do sistema de segurança de pressão. Os operadores lógicos introduzem uma complexidade às estruturas condicionais, uma vez que a combinação de múltiplas condições amplia as possibilidades de uso.





## **2.3.5 Operador lógico *or***

```python
# Usando o 'or'
a = 10
b = -1
if a > 0 or b > 0:
    print("Pelo menos uma das condições é verdadeira")

# Tpressure = 51 #bar
Tpressure1 = int(input("#Medidor 1: Qual o valor da pressão no tanque em bar? "))
Tpressure2 = int(input("#Medidor 2: Qual o valor da pressão no tanque em bar? "))
Tpressure_max = 50 # em bar

if Tpressure1 >= 50 or Tpressure2 >= 50:
  print("\n O valor da pressão do tanque está fora da pressão máxima permitida!"
        "\n Acionar protocolo de evacuação da área!")
elif Tpressure1 > 45 or Tpressure2 > 45:
  print("\n O valor da pressão do tanque está dentro da pressão máxima permitida")
  print("\n Entretanto, a pressão está na faixa de pressão que requer uma atenção")
  print("\n Mantenham o bom trabalho!")
else:
  print("\n O valor da pressão do tanque está dentro da pressão máxima permitida")
  print("\n Mantenham o bom trabalho!")
```

## **2.3.6 Operador lógico *and***

```python
# Usando o 'and'
a = 10
b = -1
if a > 0 and b > 0:
    print("As duas condições são verdadeira")

# Tpressure = 51 #bar
Tpressure1 = int(input("#Medidor 1: Qual o valor da pressão no tanque em bar? "))
Tpressure2 = int(input("#Medidor 2: Qual o valor da pressão no tanque em bar? "))
Tpressure_max = 50 # em bar

if Tpressure1 >= 50 and Tpressure2 >= 50:
  print("\n O valor da pressão do tanque está fora da pressão máxima permitida!"
        "\n Acionar protocolo de evacuação da área!")
elif Tpressure1 >= 45 or Tpressure2 >= 45:
  print("\n O valor da pressão do tanque está dentro da pressão máxima permitida")
  print("\n Entretanto, a pressão está na faixa de pressão que requer uma atenção")
  print("\n Mantenham o bom trabalho!")
else:
  print("\n O valor da pressão do tanque está dentro da pressão máxima permitida")
  print("\n Mantenham o bom trabalho!")
```

## **2.3.6 Operador lógico *in***

```python
# Lista de nutrientes no tanque
nutrientes_no_tanque = ['nitrogênio', 'fosfato', 'potássio', 'cálcio']

# Nutriente desejado para a formulação do fertilizante
nutriente_desejado = 'potássio'

# Verifica se o nutriente desejado está no tanque antes de formular o fertilizante
if nutriente_desejado in nutrientes_no_tanque:
    print(f'O nutriente {nutriente_desejado} está presente no tanque. Iniciando a formulação do fertilizante.')
else:
    print(f'O nutriente {nutriente_desejado} não está presente no tanque. Não é possível formular o fertilizante.')
```
